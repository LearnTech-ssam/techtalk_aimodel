## 2.5. Feature scaling

As seen in section 2.3.1, our features have different minimums and maximums and some features have much larger values than other features. 

>Feature Scaling is a technique to standardize the independent features present in the data in a fixed range. It is performed during the data pre-processing to handle highly varying magnitudes or values or units. If feature scaling is not done, then a machine learning algorithm tends to weigh greater values, higher and consider smaller values as the lower values, regardless of the unit of the values.

Feature scaling is beneficial because:
>- Scaling guarantees that all features are on a comparable scale and have comparable ranges...
>- Algorithm performance improvement: When the features are scaled, several machine learning methods... perform b|etter or converge more quickly...
>- Preventing numerical instability:... having features with radically differing scales can result in numerical overflow or underflow problems..
>- Scaling features makes ensuring that each characteristic is given the same consideration during the learning process. Without scaling, bigger scale features could dominate the learning, producing skewed outcomes. This bias is removed through scaling, which also guarantees that each feature contributes fairly to model predictions.

Source: https://www.geeksforgeeks.org/ml-feature-scaling-part-2/


There are multiple methods to scale features such as StandardScaler (z-score normalisation), MinMaxScaler and RobustScaler. Theoretically, the most appropriate method will depend on the distribution or shape of the data and whether there are outliers.

However,
>In practice we often ignore the shape of the distribution and just transform the data to center it by removing the mean value of each feature, then scale it by dividing non-constant features by their standard deviation.

More information can be found through the link https://scikit-learn.org/1.5/modules/preprocessing.html

We will use the RobustScaler in this instance as the method is less sensitive to outliers than MinMaxScaler.

> This Scaler removes the median and scales the data according to the quantile range (defaults to IQR: Interquartile Range). The IQR is the range between the 1st quartile (25th quantile) and the 3rd quartile (75th quantile).

Source: https://scikit-learn.org/dev/modules/generated/sklearn.preprocessing.RobustScaler.html

$$
x_{robust scaled} = \frac{x_{old} - median}{\text{75th percentile} - \text{25th percentile}}\
$$

$$
\text{interquartile range} = \text{75th percentile} - \text{25th percentile}
$$

```python
# use the x_train data to create the formula to scale x
robust_scaler = RobustScaler().fit(x_train)

# this prints out the medians and the interquartile ranges for the respective columns
print(f"""columns:
{robust_scaler.feature_names_in_.astype(str)}

medians:
{robust_scaler.center_.astype(str)}

interquartile ranges:
{robust_scaler.scale_.astype(str)}
""")
```

```python
# use the scaling formula to scale the training set
x_train_robust_scaled = robust_scaler.transform(x_train)

# it is in array format
print(f"""array format:
{x_train_robust_scaled}
""")

# we're going to add the column headers back to the data
x_train_robust_scaled = pd.DataFrame(x_train_robust_scaled, columns = robust_scaler.feature_names_in_)

print("pandas dataframe format:")
x_train_robust_scaled.head(3)
```

```python
# lets have a look at the boxplots of the scaled features
boxplots = x_train_robust_scaled.boxplot( # x is the data about the features and we are creating a boxplot for each feature
                     figsize= (10, 6), # this sets the (width, height) of the chart in inches so 10 inches by 5 inches
                     grid=False, # this removes the gridlines. Try setting this to True and rerunning it to see the gridlines
                     rot=90, # this rotates the x axis labels by 90 degrees so you can read them
                     fontsize=12 # this changes the font size to size 12
                     );

plt.tight_layout() # Adjust the padding between and around subplots by default value of 1.08. The padding is equal to 1.08 multiplied by the font size.
plt.show() # this forces the last plot to show. Useful if you have multiple plots in one code cell.
```

There are many different algorithms you can use to create a baseline model. We are doing a binary classification so would need a model that could do classification. More information on choosing a suitable algorithm can be found here https://www.geeksforgeeks.org/choosing-a-suitable-machine-learning-algorithm/

# 4. Model explanation

As well as explaining mathematically how the model works, we also want to explain how the model uses the features to predict labels to provide an explanation to colleagues.

>Coefficients in Linear SVM:
>
>In a linear SVM, each feature is assigned a coefficient that represents its importance in the decision-making process. The coefficients are the weights assigned to the features, and the magnitude of these coefficients indicates the influence of each feature on the classification decision.

Source: https://www.geeksforgeeks.org/determining-feature-importance-in-svm-classifiers-with-scikit-learn/

A bar chart visualisation and a table of the coefficients will be shown.

In the event that we have a blank or null row or all of the values in a row are zero, then we take the intercept value as the final value to consider in the algorithm.
- If this value is >=0 then the predicted label is 1.
- If this value is <0 then the predicted label is 0.

Remember, the targets we are using are 0 and 1 which represents whether the tumour associated with the image is cancerous (=0) or non-cancerous (=1).

```python
#this shows the intercept of the model
print(f"""intercept: {linear_svc_model.intercept_}
      """)

#if the intercept is <0 then predict 0, otherwise predict 1
print("In the event that we have a blank or null row or all of the values in a row are zero:")
print("predict 0, cancerous") if linear_svc_model.intercept_<0 else print("predict 1, non-cancerous")
```

```python
# this shows the coefficients of the model
print(f"""coefficients:
{linear_svc_model.coef_}
""")

# this shows the respective feature name for each coefficient
print(f"""respective feature names:
{x.columns}""")
```

```python
# to create a bar chart of the coefficients, we need to restructure the data
# we want a column for "feature_name" and a column for "coefficient"
feat_importances = pd.DataFrame(columns=["feature_name", "coefficient"])

# we want to fill the feature_name column with the feature names
feat_importances["feature_name"] = x.columns

# we want to fill the coefficient column with the model coefficients
feat_importances["coefficient"] = linear_svc_model.coef_[0]

# we want to sort the data from low to high so the bar chart is easier to read
feat_importances = feat_importances.sort_values(by="coefficient", ascending=False)

# this plots the bar chart
feat_importances.plot.barh( # we want to plot a horizontal bar chart
                           x='feature_name', # we want a bar for each feature_name
                           y='coefficient', # the values of the bar will be the coefficient
                           color=(feat_importances.loc[:,'coefficient']>=0).map({True:'green', False:'darkorange'}), # bar colour. green if coefficient is >=0 and orange if <0
                           legend=False, #remove legend
                           title="""Linear SVC Coefficients of Robust Scaled Features.
                           
                           Negative values are orange.
                           Positive values are green.""") #this is the title of the chart
# this shows the chart
plt.show()
```
