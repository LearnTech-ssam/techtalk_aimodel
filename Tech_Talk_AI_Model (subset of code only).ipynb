# Breast Cancer Diagnosis AI Model (subset of code only)

```python
#7 lines of code

import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from sklearn.svm import LinearSVC
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
#10 lines of code

chosen_random_state = 2541

#load data
x,y = load_breast_cancer(return_X_y=True, as_frame=True)

#1.Train/test split of data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=chosen_random_state)

#2. Feature manipulation
robust_scaler = RobustScaler().fit(x_train) #calculate feature scaling formula/algorithm
x_train_robust_scaled = robust_scaler.transform(x_train) #scale train data
x_test_robust_scaled = robust_scaler.transform(x_test) #scale test data

#3. Train using algorithm (mathematical model)
linear_svc_model = LinearSVC(random_state=chosen_random_state).fit(x_train_robust_scaled, y_train)

#4. Evaluate model on test data
ConfusionMatrixDisplay.from_estimator(linear_svc_model, x_test_robust_scaled, y_test);

#5. Explain how the model works
print(f"""coefficients: {linear_svc_model.coef_}""") #this shows the coefficients of the model
print(f"""respective feature names: {x.columns}""") #this shows the respective feature name for each coefficient
```
